#### Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
prod_train, prod_test = train_test_split(prod_df_model, test_size=0.33,
                                               random_state=1, shuffle=False)
```

```python
y_test = prod_test['y']
```

#### Metrics

```python
from sklearn.metrics import mean_squared_error as mse
```

#### Linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lnr = LinearRegression().fit(prod_train.iloc[:,1:].values, prod_train['y'])
```

```python
y_h_test_lnr = lnr.predict(prod_test.iloc[:,1:].values)
```

```python
rmse_lnr = np.sqrt(mse(y_test, y_h_test_lnr))/y_test.mean()
```

```python
print("NRMSE: {0:.2%}".format(rmse_lnr))
```

```python
sns.lineplot(prod_test['y'].index, y_test);
sns.lineplot(prod_test['y'].index, y_h_test_lnr, color='red');
```

#### Random forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor().fit(prod_train.iloc[:,1:].values, prod_train['y'])
```

```python
y_h_test_rfr = rfr.predict(prod_test.iloc[:,1:].values)
```

```python
rmse_rfr = np.sqrt(mse(y_test, y_h_test_rfr))/y_test.mean()
```

```python
print("NRMSE: {0:.2%}".format(rmse_rfr))
```

```python
sns.lineplot(prod_test['y'].index, y_test);
sns.lineplot(prod_test['y'].index, y_h_test_rfr, color='yellow');
```

#### XGB

```python
from xgboost import XGBRegressor
xgb = XGBRegressor().fit(prod_train.iloc[:,1:].values, prod_train['y'])
```

```python
y_h_test_xgb = xgb.predict(prod_test.iloc[:,1:].values)
```

```python
rmse_xgb = np.sqrt(mse(y_test, y_h_test_xgb))/y_test.mean()
```

```python
print("NRMSE: {0:.2%}".format(rmse_xgb))
```

```python
sns.lineplot(prod_test['y'].index, y_test);
sns.lineplot(prod_test['y'].index, y_h_test_xgb, color='green');
```

#### ARMA

```python
from statsmodels.tsa.arima_model import ARMA
results_ARMA = ARMA(endog=prod_train['y'], order =(0,1), exog=prod_train.iloc[:,1:]).fit()
```

```python
results_ARMA.summary()
```

```python
y_h_test_arma = results_ARMA.predict(start=prod_test['y'].head(1).index[0], 
                     end=prod_test['y'].tail(1).index[0],
                     exog=prod_test.iloc[:,1:])
```

```python
rmse_arma = np.sqrt(mse(y_test, y_h_test_arma))/y_test.mean()
```

```python
print("NRMSE: {0:.2%}".format(rmse_arma))
```

```python
sns.lineplot(prod_test['y'].index, y_test);
sns.lineplot(prod_test['y'].index, y_h_test_arma, color='red');
```

#### Neural network

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout

nn = Sequential()
nn.add(Dense(32, input_dim=prod_test.iloc[:,1:].shape[1], activation='relu'))
nn.add(Dropout(0.2))
nn.add(Dense(16, activation='relu'))
nn.add(Dropout(0.2))
nn.add(Dense(1, activation='linear'))
```

```python
nn.compile(loss='mean_squared_error', optimizer='adam', metrics=['mean_squared_error'])
nn.summary()
```

```python
nn.fit(prod_train.iloc[:,1:].values, prod_train['y'], epochs=15, batch_size=32, validation_split = 0.2)
```

```python
y_h_test_dnn = nn.predict(prod_test.iloc[:,1:].values).flatten()
```

```python
sns.lineplot(prod_test['y'].index, y_test);
sns.lineplot(prod_test['y'].index, y_h_test_dnn, color='black');
```

```python
rmse_dnn = np.sqrt(mse(y_test, y_h_test_dnn))/y_test.mean()
```

```python
print("NRMSE: {0:.2%}".format(rmse_dnn))
```
